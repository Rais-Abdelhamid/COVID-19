```python
plt.figure(figsize=(10,6))
data.groupby('WHO Region')['Confirmed'].sum().sort_values(ascending= False).plot(kind='bar')
plt.title('total confirmed cases by region')
plt.xlabel('region')
plt.ylabel('Total confirmed cases')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
data.groupby('Country/Region')['Active'].sum().sort_values(ascending= False).head(10). plot(kind='bar')
plt.title('Top 10 contry with highest active cases')
plt.xlabel('country')
plt.ylabel('active cases')
plt.grid(True)  # lkhto
plt.xticks(rotation=45)
plt.show()
```

```python
correlation = data[['Confirmed', 'Deaths', 'Recovered', 'Active']].corr()
correlation
```

```python
correlation.columns
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm',fmt='2f',linewidths=.5)
plt.title('correlation hatemap')
plt.show()
```

```python
data.insert(15,'fatality_rate', data.Deaths / data.Confirmed * 100)
```

```python
data['recovery_rate'] = data['Recovered'] / data['Confirmed'] * 100
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(x='fatality_rate',y='recovery_rate', data = data)
plt.title('Case fatality vs recovered')
plt.xlabel('fatality')
plt.ylabel('recovered')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
data.groupby('Country/Region')['New cases'].sum().sort_values(ascending= False).head(10). plot(kind='pie')
plt.title('Top 10 contry with highest New cases cases')
plt.xlabel('country')
plt.ylabel('New cases')
plt.grid(True)
plt.show()
```
